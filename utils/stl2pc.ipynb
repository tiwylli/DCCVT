```python
import kaolin
import numpy as np
import os

file_path = '../data/eabxhqb82fxu'
samples = 90000
mesh = kaolin.io.mesh.import_mesh(file_path+'.obj',triangulate=True)
mesh.to_batched()
pointcloud = kaolin.ops.mesh.sample_points(mesh.vertices, mesh.faces, samples)

#reshape pointcloud for export
pc = np.reshape(pointcloud[0], (samples, 3))
#insert noise to point using gaussian noise with zero mean of stadard deviation 0.005
pc = pc + np.random.normal(-0.025, 0.025, pc.shape)

    
kaolin.io.usd.export_pointcloud(file_path+'.usd', pc)

#print(pointcloud)
```
